# ML & MlOps Engineer

это инженерная дисциплина, целью которой является унификация разработки (dev) и развертывания (ops) систем машинного обучения, чтобы стандартизировать и оптимизировать непрерывную доставку высокопроизводительных моделей в производство

Задачи и инструменты:

1. Управление версиями - DVC или GiT LFS
2. Автоматизация и оркестрация конвейеров - Apache Airflow, Kubeflow Pipelines или облачных сервисов AWS Step Functions
3. Непрерывные процессы, которые расширяют классические CI/CD. К привычной непрерывной интеграции (CI) и доставке (CD) добавляются непрерывное обучение (CT) и мониторинг (CM). Модель регулярно проверяется и переобучается, если её точность падает.
4. Мониторинг в продакшне, который отслеживает здоровье модели. Особое внимание уделяется обнаружению дрейфа данных — ситуации, когда реальные данные начинают отличаться от обучающей выборки. Для этого используют специальные инструменты вроде Evidently AI или встроенные решения облачных платформ.
5. Управление экспериментами. Системы вроде MLflow или Neptune.ai автоматически записывают параметры запуска, метрики качества и результаты каждого эксперимента. Это создаёт базу знаний и упрощает сравнение разных подходов.
6. Feature Store
7. Седьмой элемент связан с развёртыванием моделей через контейнеризацию (Docker) и оркестрацию (Kubernetes). Docker and Kubernetes
